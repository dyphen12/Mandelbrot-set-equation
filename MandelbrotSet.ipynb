```python
def is_in_mandelbrot_set(c, max_iterations=100000, threshold=2):
    z = 0
    n = 0
    while abs(z) <= threshold and n < max_iterations:
        z = z*z + c
        n += 1
    return n == max_iterations
```

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot_set(xmin, xmax, ymin, ymax, width, height, max_iterations=100, threshold=2):
    x, y = np.linspace(xmin, xmax, width), np.linspace(ymin, ymax, height)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y
    C = Z
    M = np.zeros((width, height))
    for i in range(width):
        for j in range(height):
            z = 0
            c = C[i, j]
            n = 0
            while abs(z) <= threshold and n < max_iterations:
                z = z*z + c
                n += 1
            M[i, j] = n
    plt.imshow(M, cmap='gray', extent=(xmin, xmax, ymin, ymax))
    plt.show()
```

```python
mandelbrot_set(-2, 1, -1.5, 1.5, 1000, 1000)
```
